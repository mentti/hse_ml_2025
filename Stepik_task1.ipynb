```python
def solution(year: int) -> bool:
    if not bool(year % 400) or (not bool(year % 4) and bool(year % 100)):
        return True
    return False

print(solution(int(input())))
```

```python
for i in range(1889, 1890, 1):
    print(i, " - ", solution(i))

```

```python
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        a, b = b, a + b
    return(a)

print(fibonacci(12))
```

```python
import string

a_string = input()
a_string_clean = ''.join(x for x in a_string if x not in string.punctuation and not x.isspace())
a_string_clean = a_string_clean.lower()

lenx = len(a_string_clean)
median = lenx//2 
addon = 1 if lenx%2 == 0 else 0
b_string = a_string_clean[:median]
c_string = a_string_clean[:median-addon:-1]

print(b_string == c_string)
print(b_string, c_string)
```
